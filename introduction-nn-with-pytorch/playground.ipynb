## Cross entropy

```python
import torch
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator
```

Cross entropy tells how different two distributions are. It is given by:

$-\sum_{i} p_{i} \log{q_i}$,

summing over all data samples $i$.

CE is zero when p aligns with q, and increases as the 

```python
p = torch.linspace(0,1,200)
q = torch.linspace(0,1,200)

grid = torch.meshgrid(p,q, indexing='ij')

CE = - grid[0] * torch.log(grid[1]) - grid[1] * torch.log(grid[0])
```

```python
plt.pcolor(p,q,CE, cmap='coolwarm')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(grid[0], grid[1], CE, cmap='coolwarm', linewidth=0, antialiased=False)
ax.set_zlim(0, 5)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```
